```python
import numpy as np
import matplotlib.pyplot as plt
data_dictionary = {}
for i in range(1, 7):
    file_path = f'ICD_{i}_2.txt'
    try:
        with open(file_path, 'r') as f:
            data = [i.split() for i in f.readlines()]
            data_dictionary[f'ICD_{i}'] = data
    except PermissionError:
        print(f"Permission denied: Cannot access '{file_path}'. Try running as admin or checking file permissions.")
    except FileNotFoundError:
        print(f"File not found: '{file_path}'. Ensure the path is correct.")
```

```python
data_dictionary['ICD_6']
```

```python
wavelength_dict = {}
intensity_filtered_dict = {}
for key in data_dictionary.keys():
    data = data_dictionary[key]
    data = np.array(data, dtype=float)
    wavelength = data[:, 0]
    wavelength_dict[key] = wavelength
    intensity_filtered = data[:, 1]
    intensity_filtered_dict[key] = intensity_filtered
intensity_no_filter = intensity_filtered_dict['ICD_6']
transmission_dict = {}
for key in intensity_filtered_dict.keys():
    transmission = np.divide(intensity_no_filter, intensity_filtered_dict[key])
    transmission_dict[key] = transmission

print("Transmission values:", transmission_dict["ICD_1"])  
```

```python

plt.plot(wavelength, transmission, label="Transmission Function", color='b', linestyle='-')

    # Labels and title
plt.xlabel("Wavelength (nm)")
plt.ylabel("Transmission")
plt.title("Transmission Function vs. Wavelength")
plt.legend()
plt.grid(True)
plt.show()

#try adjusting the bin sizing to maybe condense the plot
```
